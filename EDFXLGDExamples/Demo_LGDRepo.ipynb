# Project Scope

Our project is complete in the sense that we have accounted for multiple situations between both EDFX and RiskCalc platforms and have attempted to address various edge cases in the event that the EDF-X API is broken or when certain Parts of the API are unresponsive (like Asynchronous requests at the endpoint level). #### This notebook provides documentation with examples for the use of the LGD Class.  We elected an object oriented approach where LGD class is also inheriting the EDFEndpoints class found within EDFXPrime.py file on the open source repo.  We have taken the liberty to provide robust functionality for users working between EDFX and RiskClac as well as for those users who wish to only use EDFX (case 3).  

#### For the users who work with both EDFX and RiskCalc the convenience the LGD class provides is tremendous.  Users now have the ability to use simply a BVDID Identifier (Case 2) or the .csv output file from the EDFX Software (Case 1 OR Case 2) and we can generate the Batch RiskCalc.csv input file with default parameters or user selected parameters within 3 lines of code for entities with existing term structures.   In additon, core LGD class functionality includes the following: 

- #### High Compute ability both within Server Side and Client Side approaches. 
    Specifically, if the endpoint within the EDFX-API has an async option (asyncResponse = True), we provide this functionality.  Withing the LGD Class this is provided on the Server Side for PDs Enpoint.  

    Regarding Client side requests, in the event that the server side requests are causing difficulties, we have written co-routines within LGD and EDFXEnpoints Classes at the mapping, pds, and LGD endpoints to increase the speed of large data requests.  We demonstrate to users the Compute challenges within this notebook.  

    An example of the compute constraint Testing is as follows:  We called pd term structures for 2594 entities from December 2017 to December 2022 with a synchronous run-time (client side basic you could say) of just under 4 hours.  Our Co-routines decreased this speed down to anywhere between 5 and 15 mintues. 

- #### Error Handling and Logging
    Specifically, if there is a case where the BVDid is not returning mapping information or the pd term structure cannot be generated the error is logged and the row is removed allowing for the contruction of the Batch RiskCalc.csv file with no errors.  This also allows us to create the LGD Term-structure with accurate payloads for Case 3

- #### Loan Scoring Flexibility
    We allow the users to build and adjust specific loan requirements in accordance to their partiuclar loan portfolio.  LossCalc Modles LGD risk at the loan level and as such, the flexibility that we provide users to adjust their specific loan information is easy to do and will help in outputting their most accurate LGD estimates relative to other data vendors.

### Important Difference for Evaluating LGD through Time in Case 3.

- If you have a loan, and you would like to see how the LGD term structure evolved you have several options. 

    You can set LGDPDEndDate to the year you want to see historically, for example, 2022-01-01 and then match that date with the LGDAsOfDate = 2022-01-01 so you will have the LGD Term structure of the entity for the 2022-01-01 period matched with the 2022-01-01 LGD Macro overlay for the same time period.
