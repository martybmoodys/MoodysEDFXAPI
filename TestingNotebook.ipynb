## Upload File
## Check Server
- This takes too long and errors out so the file i'm using is too large. 

```python
MAX_RETRIES = 30  # or however many retries you think is appropriate

# If a model input process was created we will start checking the process status every 20 seconds until it completes
if model_inputs_response:
    retries = 0
    while retries < MAX_RETRIES:
        model_inputs_process_status = endpoints.EDFXModelInputsGetStatus(processID=model_inputs_response['processId'])
        if model_inputs_process_status is None:
            print('No process status returned.')
            break
        if model_inputs_process_status['status'] == 'Processing':
            print('Model is still processing, retrying in 1 minute')
            time.sleep(60)  # wait 60 seconds and try again
            retries += 1
        else:
            # This seems to be erroring out/not receiving the file. 
            print('Model processing finished, getting files')
            files = endpoints.EDFXModelInputsGetFiles(processID=model_inputs_response['processId'])
            print(files.text)
            break
    else:
        print(f"Max retries ({MAX_RETRIES}) reached. Exiting loop.")
else:
    print('No model input response')
```

# Using a process id we already created to test without having to wait

```python
process_id = 'b675bf9a-5c6d-42bf-ae94-747b47167e51'
status = endpoints.EDFXModelInputsGetStatus(processID=process_id)
print(f'Status: {status}')
# If the process status is not "Processing" try to download the file.
# NOTE: We assume the downloaded file format is CSV, this may not be true
if status and status['status'] != 'Processing':
    print('Model processing finished, getting files')
    files = endpoints.EDFXModelInputsGetFiles(processID=process_id)
    print(files)
    with open(f'{process_id}.csv', 'w') as f:
        f.write(files)
else:
    print('Try again later')
```
